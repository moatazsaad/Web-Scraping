# **Employees data collection**

```python
import requests
import csv 
```

```python
url = 'https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy'
```

```python
employees
```

```python
employees['data'] 
```

```python
with open('employees.csv', 'w') as f:
  writer = csv.DictWriter(f, fieldnames=['name', 'email', 'gender', 'status'])
  writer.writeheader()
  for employee in employees['data']:
    name = employee['name']
    email = employee['email']
    gender = employee['gender']
    status = employee['status']
    writer.writerow({'name': name, 'email': email, 'gender': gender, 'status':status})


```

# **Corona virus**

```python
import requests
import csv 
```

```python
url = 'https://coronavirus.m.pipedream.net/'
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
data['rawData'][0]
```

```python
data['rawData'][1]
```

```python
with open('covid.csv', 'w') as f:
  writer = csv.DictWriter(f, fieldnames=['country', 'deaths', 'confirmed'])
  writer.writeheader()
  for country in data['rawData']:
    country_name = country['Country_Region']
    deaths = country['Deaths']
    confirmed = country['Confirmed']
    writer.writerow({
        'country': country_name,
        'deaths': deaths,
        'confirmed': confirmed
    })
```
